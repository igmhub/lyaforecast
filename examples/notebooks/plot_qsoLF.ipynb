# Plot Quasar LF

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import qso_LF as qLF
```

```python
# create QLF object
file1='../data/nzr_qso.dat'
# this is an older file with g-mag QLF
file2='../data/dNdzdg_QSO.dat'
LF1 = qLF.QuasarLF(file1)
LF2 = qLF.QuasarLF(file2)
mmin,mmax = LF1.range_mag()
#zmin,zmax = LF.QL_range_z()
m = np.linspace(mmin,mmax,100)
```

```python
# make plots at several z
plt.figure(figsize=(12,6))
zs = [2.0,2.5,3.0,3.5,4.0]
colors = cm.rainbow(np.linspace(0, 1, len(zs)))
for z, c in zip(zs, colors):
    plt.semilogy(m,LF1.dNdzdmddeg2(z,m)[0],color=c,ls='-',label='z='+str(z))
    plt.semilogy(m,LF2.dNdzdmddeg2(z,m)[0],color=c,ls=':')

#plt.semilogy(m,LF2.dNdzdmddeg2(2.5,m)[0],label='z=2.5')
#plt.semilogy(m,LF2.dNdzdmddeg2(3.0,m)[0],label='z=3.0')
#plt.semilogy(m,LF2.dNdzdmddeg2(3.5,m)[0],label='z=3.5')
#plt.semilogy(m,LF2.dNdzdmddeg2(4.0,m)[0],label='z=4.0')
plt.legend(loc=2,fontsize=15)
plt.ylim(1e-2,1e2)
plt.xlim(mmin,mmax)
plt.xlabel('g mag',fontsize=15)
plt.ylabel(r'dN / dz / dmag / $\rm{ddeg}^2$',fontsize=15)
plt.show()
```
