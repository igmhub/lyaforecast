# Compare bias/beta used in this code versus bias/beta in C++ version (from McDonald 2003)

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

### values assumed in python code (this code)

```python
bias=-np.sqrt(0.0173)
beta=1.58
alpha=2.9 # redshift evolution of bias
```

```python
# compute P3D at z=3, in the parallel (mu=0) and transverse (mu=1) directions
z=np.arange(2.0,4.1,0.1)
b=bias*((1+z)/(1+2.25))**alpha
```

### read beta/bias used in c++ code from simulations in McDonald (2003)

```python
bias_file='../data/z_beta_bias.txt'
inz,inbeta,inbias = np.loadtxt(bias_file,unpack=True)
```

```python
plt.plot(inz,-inbias,label='c++')
plt.plot(z,b,label='python (alpha=2.9)')
plt.plot(z,bias*((1+z)/(1+2.25))**1.25,ls=':',label='python (alpha=1.25)')
plt.legend()
plt.xlabel('z')
plt.ylabel(r'$b_\delta$')
plt.title('Linear bias as a function of redshift')
plt.savefig('bias_z.png')
```

```python
plt.plot(inz,inbeta,label='c++')
plt.plot(z,beta*np.ones_like(z),label='python (alpha=0)')
plt.plot(z,1.02*beta*((1+z)/(1+2.25))**0.75,ls=':',label='python (alpha=0.75)')
plt.legend()
plt.xlabel('z')
plt.ylabel(r'$\beta$')
plt.title('Linear RSD beta as a function of redshift')
plt.savefig('beta_z.png')
```
