```python
cosmo = CosmoCamb('Planck18.ini',2.33)
```

```python
# create theory object
theory=PowerSpectrum(cosmo)
```

```python
# compute P3D at z=2.33, in the parallel (mu=0) and transverse (mu=1) directions
z=2.33
k=np.logspace(-4,0.9,1000)
# pk_0=theory.compute_p3d_hmpc(z,k,np.zeros_like(k),-4,0.9)
# pk_1=theory.compute_p3d_hmpc(z,k,np.ones_like(k),-4,0.9)
# compute also P3D ignoring non-linear corrections (Kaiser only)
pk_0_lin=theory.compute_p3d_hmpc(z,k,np.zeros_like(k),1e-4,10,linear=True)
pk_1_lin=theory.compute_p3d_hmpc(z,k,np.ones_like(k),1e-4,10,linear=True)
```

```python
plt.ylim(1e0,2e3)
plt.xlim(1e-3,1e0)
plt.xlabel(r'k [h $\rm{Mpc}^{-1}$]',fontsize=15)
plt.ylabel(r'$\rm{P}_{\rm{F}}(k)$ [$\rm{Mpc}^3 \rm{h}^{-3}$]',fontsize=15)
plt.title(r'3D flux power at z='+str(z),fontsize=15)
# plt.loglog(k,pk_1,color='red',ls='-',label=r'$\mu=1$')
# plt.loglog(k,pk_0,color='blue',ls='-',label=r'$\mu=0$')
plt.loglog(k,pk_1_lin,color='red',ls=':',label=r'$\mu=1$')
plt.loglog(k,pk_0_lin,color='blue',ls=':',label=r'$\mu=0$')
#plt.loglog([],[],color='black',ls='-',label='non-linear')
#plt.loglog([],[],color='black',ls=':',label='linear')
plt.legend(loc='lower left',fontsize=12)
plt.show()
```

```python
p1d = theory.compute_p1d_hmpc(z,k,0.01,0.01)
p1d_TH = theory.compute_p1d_hmpc(z,k,0.01,1.0)
p1d_G = theory.compute_p1d_hmpc(z,k,1.0,0.01)
```

```python
plt.ylim(1e-5,1)
plt.xlabel(r'k [h $\rm{Mpc}^{-1}$]',fontsize=15)
plt.ylabel(r'k $\rm{P}^{1D}_{\rm{F}}(k) / \pi $',fontsize=15)
plt.title(r'1D flux power at z='+str(z),fontsize=15)
plt.loglog(k,k*p1d/np.pi,label='no smoothing')
plt.loglog(k,k*p1d_TH/np.pi,label='Gaussian smoothing')
plt.loglog(k,k*p1d_G/np.pi,label='Top hat smoothing')
plt.legend(loc='upper left',fontsize=12)
```
